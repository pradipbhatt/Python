```python
import pandas as pd
import numpy as np
```

```python
my_series = [5,8,9,10]
my_var = pd.Series(my_series)
```

```python
my_var
```

```python
#Print a specific Value
my_var[3]   #pass the index of the element as the parameter to the variable.
```

```python
# Creating labels using the index argument
my_index = ["A", 'B','C','D']
my_var = pd.Series(my_series, my_index)  #if maile my_index custom index banyera pass nagreko vaye default Index authyo.
print(my_var)

my_var["A"]  #Accessig the value using the cutom index
```

```python
cars = {
    "car1": "BMW",
    "car2": "Ford",
    "car3": "Honda",
    "Price":1212
}
my_var = pd.Series(cars)  #when we pas the dictionary to the .Series method
print(my_var)

print(f"The name of car 3 is = {my_var.loc['car3']}") #Accessing the value using the .loc method
print(f"The price of car 3 is = {my_var[2]}")    #Accessing the value using the index
```

# Video 5 How to count the data in the dataframe

```python
# Count the alag alag values lae ghatdo karam ma rakhne
my_def2["Company"].value_counts()
```

```python
# Count the alag alag values lae increasing karam ma rakhne
my_def2["Company"].value_counts(ascending=True)
```

```python
# Count Distinct Values in descending order
my_def2["Company"].value_counts()
```

```python
# Customers with the NAN
my_def2["Company"].value_counts(dropna=False)
```

```python
# Get relative frequency - Percentage
my_def2["Company"].value_counts(normalize=True)
```

```python
# Get specific Item count
my_def2["Company"].value_counts().get("Simon LLC")
my_def2["Company"].value_counts()["Simon LLC"]
```

```python
# Count Unique values- sizes
my_def2.groupby("Company").size()  #returns the unique elements of the column in the ascending order
```
